```python
# Clase Curso, atributos, nombre, duracion, costo
# funciones leerdatos(), mostrardatos()
# crear 2 objetos curso1, curso2

class Curso:
    def __init__(self):
        self.nombre = ""
        self.duracion = ""
        self.costo = ""

    def leerdatos(self):
        self.nombre = input("Ingrese el nombre del curso: ")
        self.duracion = input("Ingrese la duracion del curso: ")
        self.costo = input("Ingrese el costo del curso: ")

    def mostrardatos(self):
        print("nombre del curso: ", self.nombre)
        print("duracion del curso: ", self.duracion)
        print("costo del curso: ", self.costo)
        return self.nombre, self.duracion, self.costo

# fin de la clase Curso

curso1 = Curso()
curso1.leerdatos()
curso1.mostrardatos()

print("-+-+-+-+-+-+-+-+")

curso2 = Curso()
curso2.leerdatos()
curso2.mostrardatos()
print("-+-+-+-+-+-+-+-+")

print("Software realizado por Cepeda Darrell NL = 10")
```
